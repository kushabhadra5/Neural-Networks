#### Need of Activation function:

**An activation function in a neural network is a non-linear function applied to the output of each neuron or layer of neurons in the network. The purpose of the activation function is to introduce non-linearity into the model, allowing the neural network to learn complex, non-linear relationships between inputs and outputs.**

Following are some activation functions:
1. Sigmoid Function: This function maps any input value to a value between 0 and 1, which can be interpreted as a probability. The sigmoid function is often used in the output layer of binary classification problems.

2. ReLU (Rectified Linear Unit): This function returns the input value if it is positive, and 0 if it is negative. The ReLU function is widely used in hidden layers because it is computationally efficient and has been shown to be effective in many types of neural networks.

3. Tanh (Hyperbolic Tangent): This function maps input values to a range between -1 and 1. It is similar to the sigmoid function, but its output is zero-centered, which can help with training.

4. Softmax: This function is commonly used in the output layer of multi-class classification problems. It maps the output of each neuron to a probability distribution over the possible classes.

Let's understand this with an example:

Suppose a blind person went to supermart to purchase apples and oranges, after purchasing he want to segregate them. So, we will develop an application containing Deep Learning Neural Network that can classify apples and oranges.

### **Input Layer**:

-  Neuron 1 in Input Layer represents the pixel values of the fruit image's width.
-  Neuron 2 in Input Layer represents the pixel values of the fruit image's height.
-  Neuron 3 in Input Layer represents the color channels (e.g., RGB values) of the fruit image.

### **Hidden Layers**:

-  Each hidden layer consists of three neurons, which learn different features of the input image through the training process.
-  Neurons in the hidden layers perform complex computations based on the weights and biases associated with them.
-  Neurons in Hidden Layer 1 capture low-level features such as edges or textures.
-  Neurons in Hidden Layer 2 capture intermediate-level features combining low-level features.
-  Neurons in Hidden Layer 3 capture high-level features that are more abstract and representative of the input image.

### **Output Layers**:

-  Output Layer 1 and Output Layer 2 each represent the probability of the input image belonging to the "Apple" or "Orange" class, respectively.

Neurons in the output layers use activation functions (e.g., softmax) to convert the computed values into probabilities.
The neuron with the highest probability indicates the predicted class of the input image.
By training this neural network on the labeled fruit image dataset, the network learns to recognize patterns and features that distinguish apples from oranges. The hidden layers gradually learn to extract and combine relevant features, enabling the network to make accurate predictions about the class of new fruit images.
